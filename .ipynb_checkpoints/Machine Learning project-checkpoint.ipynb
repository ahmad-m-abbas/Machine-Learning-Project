```python
import math
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from tabulate import tabulate
from utils import mnist_reader
from sklearnex import patch_sklearn 
patch_sklearn()
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split, ParameterGrid
from sklearn.metrics import classification_report, precision_recall_fscore_support
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from tensorflow import keras
from skimage import feature as ft
import time
```

# Preparing Data
Preparing data for the Fashion MNIST dataset involves several steps, including importing the dataset into memory, splitting the data into training and testing sets, preprocessing the data, and augmenting the data. The dataset can be downloaded from the official website and loaded into memory using a library such as NumPy or TensorFlow. Once the data is loaded, it can be split into training and testing sets using a library such as scikit-learn's train_test_split function. Preprocessing the data involves normalizing or scaling the pixel values. Once the data is prepared, it can be used to train and test machine learning models.

## Reading dataset using mnist_reader library and cleaning
We will read the data using the library utils/mnist_reader which came with dataset repository on github. Hopefully, the library reads the data and converts it into NumPy array with splitting it into training and testing data. And we will shuffle the data using sklearn `shuffle`.
### Reading data

```python
X_train, y_train = mnist_reader.load_mnist('data/fashion', kind='train')
X_test, y_test = mnist_reader.load_mnist('data/fashion', kind='t10k')
X_train, y_train = shuffle(X_train, y_train)
```

### Cleaning data
Here we will check missing data. It's also worth noting that the Fashion MNIST dataset is already preprocessed and doesn't need much cleaning, but it's always a good practice to check the data for missing values, outliers, duplicates, and normalize the data.

```python
# Check for missing values
if np.isnan(X_train).sum() != 0:
    X_train[np.isnan(X_train)] = 0
```

```python
# remove duplicates
new_data = np.unique(np.insert(X_train,784, y_train, axis=1), axis=0)
X_train, y_train = new_data[:,:784], new_data[:,784:]
X_train, y_train = shuffle(X_train, y_train)
```

### Data Preprocessing
Here, we will normalize pixels values to range them between 0 and 1 using min max scaler.

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
numberToName = {0: "T-shirt", 1: "Trouser", 2: "Pullover", 3: "Dress", 4: "Coat", 5: "Sandal", 6: "Shirt", 7: "Sneaker",
                8: "Bag", 9: "Ankle boot"}
y_train = np.vectorize(numberToName.get)(y_train)
y_test = np.vectorize(numberToName.get)(y_test)
```

## Visualization 

```python
fig, axs = plt.subplots(5, 5, figsize=(8, 8))
axs = axs.ravel()
for i in np.arange(0, 25):
    axs[i].imshow(X_train[i].reshape(28,28), cmap='binary')
    axs[i].set_title(y_train[i][0])
    axs[i].axis('off')
```

### Using Principal Component Analysis
PCA is a statistical method that helps in reducing the number of dimensions in a dataset while maintaining most of its information. It is widely used for dimension reduction, particularly for high-dimensional data. This method works by calculating the correlations between dimensions and identifying a minimum number of variables that preserve the maximum amount of variation in the data. Essentially, it discards dimensions with lower explained variance and focuses on the more significant ones.

```python
features = ['pixel' + str(i+1) for i in range(X_train.shape[1]) ]
pca_df = pd.DataFrame(X_train, columns=features)
target = pd.DataFrame(y_train, columns=['label'])

pca_df['result'] = target['label']
# initializing the pca
# Set n_components as 784 dimensions
pca = PCA(n_components = 784)
# Fit on training data
pca_784 = pca.fit_transform(pca_df[features].values)
pca_df['First Dimension'] = pca_784[:,0]
pca_df['Second Dimension'] = pca_784[:,1] 
```

```python
plt.figure(figsize=(16,10))

sns.scatterplot(
    x="First Dimension", y="Second Dimension",
    hue = "result",
    hue_order = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot'],
    palette=sns.color_palette("hls", 10),
    data=pca_df,
    legend="full",
    alpha=0.3
);
```

## 1. Split Data to training and validation sets 

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=1/3, random_state=42)
```

## 2.  Features extraction
We already have the data with 28x28 feature vector from reading it

```python
head = ['X_train examples', 'X_val examples','# of Features']
all_data = [[len(X_train),len(X_val),len(X_train[0])]]
print(tabulate(all_data, headers=head, tablefmt="pipe"))
```

## 3. Baseline model
As a baseline model, we will evaluate a nearest neighbor baseline. For the baseline, we have decided
two distances of our choice, euclidian and manhatten distances, and we set k to either 1 or 3. In total, we  have 
4 different combinations of the distance measure and k. we choose the best combination by evaluating
all our choices on the validation set.

```python
head = ['k','metric','p','score in  validation set']
data =[]

knn_param_grid = {'n_neighbors': [1,3], 'metric': ['minkowski'], 'p': [1,2]}
knn_grid = ParameterGrid(knn_param_grid)

for param in knn_grid:
    knn_model = KNeighborsClassifier(**param, algorithm='brute')  # unpacking param which is a dictionary
    knn_model.fit(X_train, y_train.ravel())
    score = knn_model.score(X_val, y_val)
    list_entry = [param['n_neighbors'],  param['metric'],  param['p'],score]
    data.insert(0,list_entry)
print(tabulate(data, headers=head, tablefmt="pipe"))
max_accuracy = max(entry[3] for entry in data)
for item in data:
    if item[3] == max_accuracy:
        best_model = item
print("The highest Accuracy %f is the model model with K = %d and used %s distance with p=%d" %
                (best_model[3],best_model[0],best_model[1],best_model[2]))
```

```python
start_time = time.time()
best_knn_model = KNeighborsClassifier(metric=best_model[1], algorithm='brute',n_neighbors=best_model[0], p=best_model[2])
best_knn_model.fit(X_train, y_train.ravel())
y_pred = best_knn_model.predict(X_test)
knn_accuracy = best_knn_model.score(X_test, y_test)
knn_f_score = precision_recall_fscore_support(y_test,y_pred, average='macro')[2]
print("--- %s seconds ---" % (time.time() - start_time))
print("The KNN model with K = %d and used %s distance has %s accuracy this classification report:"
      % (best_model[0],best_model[1], knn_accuracy))
knn_report = classification_report(y_test,y_pred,digits = 5)
print(knn_report)
```

```python
con_matrix = pd.crosstab(pd.Series(pd.DataFrame(y_test).values.flatten(), name='Actual' ),pd.Series(pd.DataFrame(y_pred).values.flatten(), name='Predicted'))
plt.figure(figsize = (9,9))
plt.title("Confusion Matrix on KNN")
sns.heatmap(con_matrix, cmap="Greys", annot=True, fmt='g')
plt.show()
```

### 4.1. SVM
This report utilizes the SVM algorithm for its memory efficiency, and examines the performance of models on the dataset. Polynomial, sgimoid, linear, and radial basis function kernel functions are considered. The optimization process combines both manual and grid search techniques to determine the optimal values for the regularization parameter C and the kernel function.

```python
head = ['C','kernel','score in validation set']
data =[]
svm_param_grid = {'C': [1, 10, 100], 'kernel': ['sigmoid', 'linear', 'poly', 'rbf']}
svm_grid = ParameterGrid(svm_param_grid)
start_time_over_all = time.time()
for param in svm_grid:
    start_time = time.time()
    svm_model = SVC(**param)  
    svm_model.fit(X_train, y_train.ravel())
    score=svm_model.score(X_val, y_val)
    list_entry = [param['C'],  param['kernel'],score]
    print("--- %s seconds for C = %d and  Kernal = %s with accuracy = %f---"
          % (time.time() - start_time, param['C'],  param['kernel'], score))
    data.insert(0,list_entry)
print(tabulate(data, headers=head, tablefmt="pipe"))
print("--- %s seconds ---" % (time.time() - start_time))
max_accuracy = max(entry[2] for entry in data)
for item in data:
    if item[2] == max_accuracy:
        best_model = item
print("The highest Accuracy %f is the model model with C = %d and used %s kernal" %
                (best_model[2],best_model[0],best_model[1]))
```

```python
start_time = time.time()
svm_model = SVC(C = 10, kernel='rbf')
svm_model.fit(X_train, y_train.ravel())
svm_accuracy=svm_model.score(X_test, y_test)
predictions = svm_model.predict(X_test)
svm_f_score = precision_recall_fscore_support(y_test,predictions, average='macro')[2]
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
con_matrix = pd.crosstab(pd.Series(pd.DataFrame(y_test).values.flatten(), name='Actual' ),pd.Series(pd.DataFrame(predictions).values.flatten(), name='Predicted'))
plt.figure(figsize = (9,9))
plt.title("Confusion Matrix on SVM")
sns.heatmap(con_matrix, cmap="Greys", annot=True, fmt='g')
plt.show()
```

### 4.2. Random Forrest
Random Forest is a popular machine learning algorithm used for classification and regression tasks. It operates by constructing multiple decision trees and combining their outputs to make a prediction. In a Jupyter notebook, Random Forest can be implemented using libraries such as scikit-learn. By using Random Forest in a Jupyter notebook, one can easily train the model on a dataset, make predictions, and visualize the results through the use of interactive plots and charts. The Jupyter notebook environment makes it easy to experiment with different model hyperparameters and evaluate the performance of the model.

```python
head = ['n_estimators','criterion','max_depth','score in validation set']
data =[]
rf_param_grid = {'n_estimators': [10, 50, 100], 'criterion': ['entropy', 'gini'],
                'max_depth': [10, 50, 100],
                'max_features': ["sqrt", "log2"]}
rf_grid = ParameterGrid(rf_param_grid)
start_time_over_all = time.time()
for param in rf_grid:
    start_time = time.time()
    rf_model = RandomForestClassifier(**param)  
    rf_model.fit(X_train, y_train.ravel())
    score=rf_model.score(X_val, y_val)
    list_entry = [param['n_estimators'],  param['criterion'], param['max_depth'], param['max_features'],score]

    data.insert(0,list_entry)
print(tabulate(data, headers=head, tablefmt="pipe"))
print("--- %s seconds ---" % (time.time() - start_time))
max_accuracy = max(entry[4] for entry in data)
for item in data:
    if item[4] == max_accuracy:
        best_model = item
print("The highest Accuracy %f is the model model with n_estimators = %d and used %s criterion with max_depth = %s and max_features splitter %s"
      % (best_model[4],best_model[0],best_model[1], best_model[2], best_model[3]))
```

It is difficult to differentiate coats, pullovers, and shirts based on their appearance alone, making Fashion-MNIST a more challenging task than MNIST. The misclassifications in the Fashion-MNIST dataset can be attributed to several reasons:

- Similar appearance: Classes such as T-shirt, shirt, pullover, and coat have similar appearances, making it difficult for the model to differentiate between them.

- Limited features: The 28x28 grayscale images in the dataset do not contain enough features to accurately distinguish between certain classes, such as shirt and t-shirt.

Several actions might be made to enhance a SVM model. Increase the diversity of training images for the shirt class: The pattern of misclassifying the shirt class as other similar classes suggests that the model may be over-fitting to the training images. Increasing the diversity of training images for the shirt class can help the model better generalize to new, unseen data.

```python
# Finding indices of all the shirt images in the training set
indices = np.where(y_train == "Shirt")[0]

# Selecting only 20% of the shirt images
selected_indices = np.random.choice(indices, size=int(0.2 * len(indices)), replace=False)

# Adding the selected shirt images to the training set
X_augmented = X_train[selected_indices]
y_augmented = y_train[selected_indices]

# Augmenting the selected images to increase the diversity of the shirt class
from keras.preprocessing.image import ImageDataGenerator

# Define data generator
datagen = ImageDataGenerator(
        featurewise_center=False, 
        samplewise_center=False,  
        featurewise_std_normalization=False,  
        samplewise_std_normalization=False,  
        zca_whitening=False,
        rotation_range=3, 
        zoom_range = 0.1,
        width_shift_range=0.1,
        height_shift_range=0.1) 

# Fit the generator to the selected images
datagen.fit(X_augmented.reshape(len(X_augmented), 28, 28, 1))

# Generate new, augmented images
for X_batch, y_batch in datagen.flow(X_augmented.reshape(len(X_augmented), 28, 28, 1), y_augmented, batch_size=32):
    X_augmented = np.concatenate((X_augmented, X_batch.reshape(len(X_batch),784)))
    y_augmented = np.concatenate((y_augmented, y_batch))
    if len(X_augmented) >= len(indices):
        break

# Adding the augmented images to the original training set
X_train = np.concatenate((X_train, X_augmented))
y_train = np.concatenate((y_train, y_augmented))
```

```python
start_time = time.time()
improved_svm_model = SVC(C = 10, kernel="rbf")
improved_svm_model.fit(X_train, y_train.ravel())
improved_svm_accuracy = improved_svm_model.score(X_test, y_test)
predictions = improved_svm_model.predict(X_test)
improved_svm_f_score = precision_recall_fscore_support(y_test,predictions, average='macro')[2]
print("--- %s seconds ---" % (time.time() - start_time))
```

Although accuracy isn't usually the ideal statistic to use, it can be a helpful indicator for assessing a model's performance in the Fashion-MNIST challenge. The percentage of accurate predictions made by the model is measured, although accuracy does not give a full view of the model's performance.

Accuracy can be deceptive in some situations, such as when the classes are unbalanced (one class has significantly more data than the others). A model that consistently predicts T-shirt, for instance, would still have a high accuracy even if it was inaccurately classifying the other classes in the Fashion-MNIST dataset. Metrics like accuracy, recall, or F1-score might offer a better insight of the model's performance under such circumstances.

The price of misclassification is another aspect to take into account. In certain situations, misclassifying a shirt as a T-shirt might not have a big impact, whereas misclassifying a coat as a pullover might. In these circumstances, it can be preferable to assess the model's performance using metrics like the confusion matrix or ROC curve.

In conclusion, accuracy can be a helpful statistic for assessing a model's performance in the Fashion-MNIST task, but it's vital to take the job's context and potential limits into account before adopting accuracy as the only metric. The performance of the model may be assessed using other measures in a more thorough manner.

## 6. Different Features
HOG are features that can be used to improve the accuracy of models in the Fashion-MNIST task. HOG captures gradient information. By combining both features, HOG and 784 pixles, the model can benefit from both gradient and texture information, leading to improved accuracy. Using different features makes the model less reliant on a single type of information, improving its robustness and reducing the risk of overfitting. 

```python
train_features = pd.DataFrame(X_train)
train_label  = pd.DataFrame(y_train)

test_features = pd.DataFrame(X_test)
test_label = pd.DataFrame(y_test)

train_features = train_features.values.reshape((-1, 28, 28))
test_features = test_features.values.reshape((-1, 28, 28))
```

```python
def HOG_img(img):
    features = ft.hog(img, orientations=9, pixels_per_cell=(4, 4),
                      cells_per_block=(2, 2), block_norm="L2")
    return features
```

```python
train_HOG_ft = []
for index, img in enumerate(train_features):
    print('\rHOG: ' + str(round((index+1)/len(train_features)*100, 1)) + '%', end='')
    train_HOG_ft.append(HOG_img(img))
print('')

test_HOG_ft = []
for index, img in enumerate(test_features):
    print('\rHOG: ' + str(round((index+1)/len(test_features)*100, 1)) + '%', end='')
    test_HOG_ft.append(HOG_img(img))
print('')
```

```python
train_HOG_ft = np.concatenate((train_HOG_ft, X_train), axis=1)
test_HOG_ft = np.concatenate((test_HOG_ft, X_test), axis=1)
```

```python
HOG_svm_model = SVC(C=10, kernel='rbf')
HOG_svm_model.fit(train_HOG_ft, train_label)
HOG_svm_accuracy = HOG_svm_model.score(test_HOG_ft, test_label)
predictions = HOG_svm_model.predict(test_HOG_ft)
HOG_svm_f_score = precision_recall_fscore_support(y_test,predictions, average='macro')[2]
HOG_svm_accuracy
```

```python
HOG_knn_model = KNeighborsClassifier(algorithm='brute', 
                           metric='minkowski',
                           p = 1 ,
                           n_neighbors=3)
HOG_knn_model.fit(train_HOG_ft, train_label)
y_pred = HOG_knn_model.predict(test_HOG_ft)
HOG_knn_accuracy = HOG_knn_model.score(test_HOG_ft, test_label)
predictions = HOG_knn_model.predict(test_HOG_ft)
HOG_knn_f_score = precision_recall_fscore_support(y_test,predictions, average='macro')[2]
HOG_knn_accuracy
```

```python
HOG_rf_model = RandomForestClassifier(n_estimators = 100, criterion  = 'entropy', max_depth =100, max_features='sqrt')
HOG_rf_model.fit(train_HOG_ft, train_label)
HOG_rf_accuracy = HOG_rf_model.score(test_HOG_ft, test_label)
predictions = HOG_rf_model.predict(test_HOG_ft)
HOG_rf_f_score = precision_recall_fscore_support(y_test,predictions, average='macro')[2]
HOG_rf_accuracy
```

```python
head = ['Model', 'Accuracy', 'F1-score']
data = []
data.insert(0, ['KNN', HOG_knn_accuracy, HOG_knn_f_score])
data.insert(0, ['SVM', HOG_svm_accuracy, HOG_svm_f_score])
data.insert(0, ['Random Forrest', HOG_rf_accuracy, HOG_rf_f_score])
print(tabulate(data, headers=head, tablefmt="pipe"))
```
