```python
import math
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from tabulate import tabulate
from utils import mnist_reader
import plotly.graph_objects as go
from sklearnex import patch_sklearn 
patch_sklearn()
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split, ParameterGrid
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import tensorflow as tf
from keras.layers import Dense, Conv2D, Activation, MaxPool2D, Flatten, Dropout, BatchNormalization
import tensorflow as tf
from tensorflow import keras
from keras.utils import plot_model
from skimage import feature as ft
import visualkeras
import time
```

```python
X_train, y_train = mnist_reader.load_mnist('data/fashion', kind='train')
X_test, y_test = mnist_reader.load_mnist('data/fashion', kind='t10k')
X_train, y_train = shuffle(X_train, y_train)
# Check for missing values
if np.isnan(X_train).sum() != 0:
    X_train[np.isnan(X_train)] = 0

# remove duplicates
new_data = np.unique(np.insert(X_train,784, y_train, axis=1), axis=0)
X_train, y_train = new_data[:,:784], new_data[:,784:]
X_train, y_train = shuffle(X_train, y_train)

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


fig, axs = plt.subplots(5, 5, figsize=(8, 8))
axs = axs.ravel()
for i in np.arange(0, 25):
    axs[i].imshow(X_train[i].reshape(28,28), cmap='binary')
    axs[i].set_title(y_train[i])
    axs[i].axis('off')
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=1/3, random_state=42)
head = ['X_train examples', 'X_val examples','# of Features']
all_data = [[len(X_train),len(X_val),len(X_train[0])]]
print(tabulate(all_data, headers=head, tablefmt="pipe"))
```

# CNN

```python
# Load the FashionMNIST dataset
x_train, y_train, x_test, y_test = X_train, y_train, X_test, y_test

# Split the training data into a training set and a validation set
x_val = X_val
y_val = y_val
# Reshape the data to be used in a CNN
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)
x_val = x_val.reshape(-1, 28, 28, 1)

# Define a set of hyperparameters to evaluate
batch_sizes = [64, 128]
num_filters = [32, 64]
kernel_sizes = [3, 5]

# Create a dictionary to store the results
results = {}

# Loop over the different combinations of hyperparameters
for batch_size in batch_sizes:
    for num_filter in num_filters:
        for kernel_size in kernel_sizes:
            # Create a model with the current hyperparameters
            model = keras.Sequential()

            model.add(keras.layers.Conv2D(num_filter, (kernel_size,kernel_size), padding='same', input_shape=(28, 28, 1)))
            model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))

            model.add(keras.layers.Conv2D(64, (3,3), padding='same', activation=tf.nn.relu))
            model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))

            model.add(keras.layers.Flatten())
            model.add(keras.layers.Dense(128, activation='relu'))
            
            model.add(keras.layers.Dense(10, activation='softmax'))
            model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

            # Train the model on the training set and evaluate its performance on the validation set
            history = model.fit(x_train, y_train, batch_size=batch_size, epochs=20, validation_data=(x_val, y_val))
            val_acc = history.history['val_accuracy'][-1]

            # Store the results in the dictionary
            results[(batch_size, num_filter, kernel_size)] = val_acc

# Find the hyperparameters that result in the best performance on the validation set
best_hyperparams = max(results, key=results.get)
```

```python
# X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=1/6, random_state=42)
train_dataset = pd.DataFrame(X_train)
train_dataset[784]=y_train

test_dataset = pd.DataFrame(X_test)
test_dataset[784]=y_test
```

```python
def data_preprocessing(raw):
    label = tf.keras.utils.to_categorical(raw[784], 10)
    num_images = raw.shape[0]
    x_as_array = raw.values[:,1:]
    x_shaped_array = x_as_array.reshape(num_images, 28, 28, 1)
    image = x_shaped_array / 255
    return image, label

X, y = data_preprocessing(train_dataset)
X_test, y_test = data_preprocessing(test_dataset)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=1/3, random_state=42)
```

```python
y_test
```

```python
best_hyperparams 
```

```python
# Train the final model on the combined training and validation sets using the selected hyperparameters
model = keras.Sequential()

model.add(keras.layers.Conv2D(best_hyperparams[1], (best_hyperparams[2], best_hyperparams[2]),padding='same', input_shape=(28, 28, 1)))
model.add(keras.layers.MaxPooling2D(2,2))

model.add(keras.layers.Conv2D(64, (3,3), padding='same', activation=tf.nn.relu))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))

model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(128, activation='relu'))

model.add(keras.layers.Dense(10, activation='softmax'))

model.compile(optimizer='adam', loss=tf.keras.losses.categorical_crossentropy, metrics=['accuracy'])

X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
X_val = X_val.reshape(X_val.shape[0], 28, 28, 1)
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)

train_model = model.fit(X_train, y_train,
                  batch_size=best_hyperparams[0],
                  epochs=100,
                  verbose=1,
                  validation_data=(X_val, y_val))
```

```python
epoch = len(train_model.history.get('loss',[]))

# Draw Model Accuracy
plt.figure(2,figsize=(6,4))
plt.plot(range(epoch),train_model.history['accuracy'])
plt.plot(range(epoch),train_model.history['val_accuracy'])
plt.xlabel('# Epochs')
plt.ylabel('Accuracy')
plt.title('Model Accuracy')
plt.grid(True)
plt.legend(['train','validation'],loc=4)
plt.style.use(['classic'])

# Draw Model Loss
plt.figure(1,figsize=(6,4))
plt.plot(range(epoch),train_model.history['loss'])
plt.plot(range(epoch),train_model.history['val_loss'])
plt.xlabel('# Epochs')
plt.ylabel('Loss')
plt.title('Model Loss')
plt.grid(True)
plt.legend(['train','validation'])
plt.style.use(['classic'])
```

```python
score = model.evaluate(X_test, y_test)
# checking the test loss and test accuracy
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
